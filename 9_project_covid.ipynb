# Data Preparation

```python
# Dropping Unwanted columns
df.drop(['Country_Region','Last_Update','Lat','Long_'],axis=1,inplace=True)
```

```python
df
```

Observation:
    
    I have dropped lattitude,longitude, country_region because there is Province_state.
    
    i have dropped last_update Because it has only one unique value

```python
# Encoding the object attribute

from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df['ISO3']=le.fit_transform(df['ISO3'])
df['Province_State']=le.fit_transform(df['Province_State'])
```

```python
df
```

```python
# check for skewness
df.skew()
```

```python
# Removing skeweness
for i in df.columns:
    if df[i].skew()>1:
        df[i]=np.sqrt(df[i])
```

```python
df.skew()
```

# Model training

```python
# I want to predict number of deaths by regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x=df.drop('Deaths',axis=1)

x
```

```python
y=df['Deaths']

y
```

```python
# standardization of data

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score
linearscore = cross_val_score(lr,x,y,cv=6)
print(linearscore)
print(linearscore.mean(),linearscore.std())
```

Obseravtion:
    
    We can see cross validation score looking good

```python
# predict the x_test

pred=lr.predict(x_test)
```

```python
print("Mean absolute error : ",mean_absolute_error(y_test,pred))   
print("Mean squared error : ",mean_squared_error(y_test,pred))
print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,pred)))
print('r2 score :',r2_score(y_test,pred))
```

# Check various for various methods using for loop

```python
# Check for various methods using for loop.
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso,Ridge,ElasticNet
import warnings
warnings.filterwarnings('ignore')


model=[SVR(),DecisionTreeRegressor(),KNeighborsRegressor(),Lasso(),Ridge(),ElasticNet()]

for m in model:
    m.fit(x_train,y_train)
    print('score of',m,'is : ',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print("Mean absolute error : ",mean_absolute_error(y_test,predm))
    print("Mean squared error : ",mean_squared_error(y_test,predm))
    print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,predm)))
    print('r2 score :',r2_score(y_test,predm))
    print('\n')      
        
```

Observation:
    
    we can observe the above all moodels we can say that Ridge regression giving best accuracy with least error and r2 score

# Ensemble

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

ensemble=[RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]

for m in ensemble:
    m.fit(x_train,y_train)
    print('score of',m,'is : ',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print("Mean absolute error : ",mean_absolute_error(y_test,predm))
    print("Mean squared error : ",mean_squared_error(y_test,predm))
    print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,predm)))
    print('r2 score :',r2_score(y_test,predm))
    print('\n')      
```

Observation:
    
    From above all observation Ridge regression giving good results 

# GridsearchCV

```python
from sklearn.model_selection import GridSearchCV

rd=Ridge()
parameters={'alpha':[0.0001,0.001,0.01,0.1,1,10,100]}
ridge=GridSearchCV(rd,parameters,scoring='neg_mean_squared_error',cv=5)
ridge.fit(x,y)
```

```python
ridge.best_params_
```

```python
rd=Ridge(alpha=0.01)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

# Saving Best results

```python
# I want to save the Ridge model 

import joblib
joblib.dump(rd,'covid_Ridge_model.ob')
```

```python
#loading saved model
ridge_model=joblib.load('covid_Ridge_model.ob')
```

```python
# predict the deaths using saved model
prediction=ridge_model.predict(x_test)
```
