# Dropping unwanted columns

```python
df_hr=df_hr.drop(['EmployeeCount','Over18','StandardHours'],axis=1)
```

```python
df_hr
```

# Encoding the Categorical features

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
```

```python
df_hr['Attrition']=le.fit_transform(df_hr['Attrition'])
df_hr['BusinessTravel']=le.fit_transform(df_hr['BusinessTravel'])
df_hr['Department']=le.fit_transform(df_hr['Department'])
df_hr['EducationField']=le.fit_transform(df_hr['EducationField'])
df_hr['Gender']=le.fit_transform(df_hr['Gender'])
df_hr['JobRole']=le.fit_transform(df_hr['JobRole'])
df_hr['MaritalStatus']=le.fit_transform(df_hr['MaritalStatus'])
df_hr['OverTime']=le.fit_transform(df_hr['OverTime'])
```

```python
df_hr
```

# Removing Outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(df_hr))
z
```

```python
thershold=3
print(np.where(z>3))
```

```python
# dropping Outliers
hr_new=df_hr[(z<3).all(axis=1)]
```

```python
hr_new
```

We can observe 83 rows of Outliers are removed from dataset

# Building a Classification model

```python
# By using LogisticRegression i want to predict Attrition because it has binary categories

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
x=hr_new.drop('Attrition',axis=1)
```

```python
x
```

```python
y=hr_new['Attrition']
```

```python
y
```

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.score(x_train,y_train)
```

```python
predlg=lg.predict(x_test)
```

```python
predlg
```

```python
# accuracy of model on test data
print('Accuracy score : ',accuracy_score(y_test,predlg))
```

We can see Accuracy  is 85%

```python
# Confussion Matrix
print(confusion_matrix(y_test,predlg))
```

```python
# Classification Report
print(classification_report(y_test,predlg))
```

# check for various  methods

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=[GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))

    print('\n')      
        
```

From Above all various methods Logisticregression giving best accuracy score

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
lgscores= cross_val_score(lg,x,y,cv=5)
print(lgscores)
print(lgscores.mean(),lgscores.std())
```

# Ensemble

```python
# RandomForestClassifier 
from sklearn.ensemble import RandomForestClassifier 


rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(x_train,y_train) 
rf.score(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

We can see RandomForestClassifier is giving 89% of accuracy

```python
# Ada Boost Classifier
from sklearn.ensemble import AdaBoostClassifier

ad=AdaBoostClassifier(n_estimators=50)
ad.fit(x_train,y_train)
ad_predict=ad.predict(x_test)
print(accuracy_score(y_test,ad_predict))
print(confusion_matrix(y_test,ad_predict))
print(classification_report(y_test,ad_predict))
```

We can see AdaBoostclassifier is giving 87% of accuracy

# saving best model

```python
# save the best model in object file
import joblib
joblib.dump(rf,'hr_analyticsRF.ob')
```
