# Machine Learning

Building Linear regression model to predict AveragePrice 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x=df_new.drop('AveragePrice',axis=1)
```

```python
y=df_new['AveragePrice']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=55)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

```python
# predict the value
pred=lm.predict(x_test)
print("predicted result price : ",pred)
print('actual price : ',y_test)
```

```python
print("Mean absolute error : ",mean_absolute_error(y_test,pred))
print("Mean squared error : ",mean_squared_error(y_test,pred))

print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
# R2 score
print(r2_score(y_test,pred))
```

Linearregression is giving low score and low Root mean square error. 

Low score results because indepentent and dependent attributes are not related linearly.

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
lmscores= cross_val_score(lm,x,y,cv=4)
print(lmscores)
print(lmscores.mean(),lmscores.std())
```

# checking for various Methods

```python
# Check for various methods using for loop.
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso,Ridge,ElasticNet
import warnings
warnings.filterwarnings('ignore')


model=[SVR(),DecisionTreeRegressor(),KNeighborsRegressor(),Lasso(),Ridge(),ElasticNet()]

for m in model:
    m.fit(x_train,y_train)
    print('score of',m,'is : ',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print("Mean absolute error : ",mean_absolute_error(y_test,predm))
    print("Mean squared error : ",mean_squared_error(y_test,predm))
    print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,predm)))
    print('r2 score :',r2_score(y_test,predm))
    print('\n')      
        
```

From observation we can see DecisiontreeRegressor giving highest score and low errors.

# Ensemble

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
print('score of is : ',rf.score(x_train,y_train))
predrf=rf.predict(x_test)
print("Mean absolute error : ",mean_absolute_error(y_test,predrf))
print("Mean squared error : ",mean_squared_error(y_test,predrf))
print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,predrf)))
print('r2 score :',r2_score(y_test,predrf))
```

```python
from sklearn.ensemble import AdaBoostRegressor
ad=AdaBoostRegressor()
ad.fit(x_train,y_train)
print('score of is : ',ad.score(x_train,y_train))
pred_ad=ad.predict(x_test)
print("Mean absolute error : ",mean_absolute_error(y_test,pred_ad))
print("Mean squared error : ",mean_squared_error(y_test,pred_ad))
print("Root mean squared error : ",np.sqrt(mean_squared_error(y_test,pred_ad)))
print('r2 score :',r2_score(y_test,pred_ad))
```

From above Observations we can see RandomForestRegressor is giving highest accuracy,low errors and good r2 score.

# Saving the best model

```python
# I want to store the RandomForestRegressor model beacause it has highest accuracy and low error

import joblib

joblib.dump(rf,'avcadorf.ob')
```

```python
# Loading the saved model
avcd=joblib.load('avcadorf.ob')
```

```python
#Predict values using saved model
predictions=avcd.predict(x_test)
predictions
```

```python
# Saving predicted values in CSV file
output=pd.DataFrame(predictions)
output.index=x_test.index
output.columns=['Predictions']
output.to_csv('avcd_predicted_result.csv')
```

conclusions:
    
    for above avacoda dataset predicting the AveragePrice by using Regression, RandomForestregressor is giving highest accuracy that is 97.3% with least errors.

# Check various methods using for loops

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')


model=[GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]

for m in model:
    m.fit(train_x,train_y)
    m.score(train_x,train_y)
    predm=m.predict(test_x)
    print('Accuracy score of',m,'is:')
    print(accuracy_score(test_y,predm))
    print(confusion_matrix(test_y,predm))
    print(classification_report(test_y,predm))

    print('\n')      
        
```

from above observation GuassianNB() is giving highest accuracy

# cross validation
